```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import json
from config import large_figure_width, large_figure_height, standard_figure_width, standard_figure_height


W = 54
w = 36
T = 2.5
B = 11.4
h = 4.5
Ac = (W+w)*h/2
Ab = B*T
Hm = Ac/W
m = Ab/Ac

p = (W-w)/2/h
g = 9.81


print('m = ', m)
print('p = ', p)
z = np.linspace(1e-4, 2, 1000)

z = z/Hm



Fc = np.sqrt( ( 2*z*(1-z-m)**2 )/( 1-(1-z-m)**2 ) ) # constantine
Fct = np.sqrt( ( 2*z*(1-z-m+p*Hm/W*z**2)**2 )/( 1-(1-z-m+p*Hm/W*z**2)**2 ) ) # constantine trapeze



# Approximationd de Kim pour u
# uct0 = z/Fct**2
# uc0 = z/Fc**2

uc = -1 + np.sqrt( 1 + 2*z/Fc**2 )
uct = -1 + np.sqrt( 1 + 2*z/Fct**2 )




max_index = np.argmax(Fc)
Fc_max = Fc[max_index]
zc_max = z[max_index]
uc_max = uc[max_index]

max_index_ct = np.argmax(Fct)
Fct_max = Fct[max_index_ct]
zct_max = z[max_index_ct]
uct_max = uct[max_index_ct]




fig_size = (1.2*large_figure_width, 1.7*large_figure_height)
fig, axes = plt.subplots(2, 1, figsize=fig_size, sharex=True)


ax1 = axes[1]

ax1.plot(Fct[:max_index_ct], z[:max_index_ct], label='KSC (étendu au cas trapèze)', color='tab:blue')
ax1.plot(Fct[max_index_ct], z[max_index_ct], 'o', color='tab:blue')
ax1.plot( Fct[max_index_ct:], z[max_index_ct:], '--',color = 'tab:blue' )

ax1.plot(Fc[:max_index], z[:max_index], label='KSC (rectangle équivalent)', color='black')
ax1.plot(Fc[max_index], z[max_index], 'o', color='black')
ax1.plot( Fc[max_index:], z[max_index:], '--',color = 'black' )

ax1.set_xlabel(r'$Fn_{Hm}$', rotation=0)
ax1.set_ylabel(r'$\dfrac{z}{H_m}$', rotation=0)
ax1.legend(loc =2)


ax2 = axes[0]


ax2.plot(Fct[:max_index_ct], uct[:max_index_ct], label='KSC (étendu au cas trapèze)', color='tab:blue')
ax2.plot(Fct[max_index_ct], uct[max_index_ct], 'o', color='tab:blue')
ax2.plot( Fct[max_index_ct:], uct[max_index_ct:], '--',color = 'tab:blue' )


ax2.plot(Fc[:max_index], uc[:max_index], label='KSC (rectangle équivalent)', color='black')
ax2.plot(Fc[max_index], uc[max_index], 'o', color='black')
ax2.plot( Fc[max_index:], uc[max_index:], '--',color = 'black' )

ax2.set_xlabel(r'$Fn_{Hm}$')
ax2.set_ylabel(r'$\dfrac{u}{V}$', rotation=0)
ax2.set_ylim( 0, 1.2 )
ax2.legend(loc =2)

ax1.yaxis.set_label_coords(-0.12, 0.45)
ax2.yaxis.set_label_coords(-0.1, 0.5)

ax2.text( x=0, y = 0.8, s = '(a)' )
ax1.text( x=0, y = 0.37, s = '(b)' )


plt.savefig( '../figures/quintique-vs-cubique.pdf', bbox_inches = 'tight')

plt.show()
```
