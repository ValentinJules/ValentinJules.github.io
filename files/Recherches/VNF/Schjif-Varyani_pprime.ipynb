```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import json
from config import large_figure_width, standard_figure_width
canal = 'Canal trapézoïdal'
W = 54
w = 8
T = 2.5
B = 6
h = 4.5
Ac = h*W
Ab = B*T
Hm = Ac/W
m = Ab/Ac

p = (W-w)/2/h
gamma = 0.9
Cd = gamma*(T/Hm)**2



print('m = ', m)
z = np.linspace(0, 2, 1000)

z = z/Hm


F = np.sqrt( ( 2*z*(1-z*(1-B/W)-m+p*Hm/W*z**2)**2 )/( 1-(1-z*(1-B/W)-m+p*Hm/W*z**2)**2 ) ) #complet

Fc = np.sqrt( ( 2*z*(1-z-m)**2 )/( 1-(1-z-m)**2 ) ) # constantine


Fv = np.sqrt( ( 2*z*(1-m-z*(1-B/W))**2 )/( 1-(1-m-z*(1-B/W))**2 ) ) #varyani

Fct = np.sqrt( ( 2*z*(1-z-m+p*Hm/W*z**2)**2 )/( 1-(1-z-m+p*Hm/W*z**2)**2 ) ) # constantine trapeze

Fb = np.sqrt( (2*z*( 1 - m ) - (1-B/W)*z**2 + (2*p*Hm)/(3*W)*z**3 )/(Cd*(m+B/W*z)-2+2*(1-z-m+p*Hm/W*z**2)**(-1)) )



peaks_indices_c, _ = find_peaks(Fc)
peaks_indices_ct, _ = find_peaks(Fct)
peaks_indices_v, _ = find_peaks(Fv)
peaks_indices, _ = find_peaks(F)
peaks_indices_b, _ = find_peaks(Fb)

# Obtenir les valeurs de z et F correspondant aux maxima locaux
zc_peaks = z[peaks_indices_c]
Fc_peaks = Fc[peaks_indices_c]

zct_peaks = z[peaks_indices_ct]
Fct_peaks = Fct[peaks_indices_ct]

zv_peaks = z[peaks_indices_v]
Fv_peaks = Fv[peaks_indices_v]

z_peaks = z[peaks_indices]
F_peaks = F[peaks_indices]

zb_peaks = z[peaks_indices_b]
Fb_peaks = Fb[peaks_indices_b]


plt.figure()

plt.plot( Fct[z<zct_peaks], z[z<zct_peaks], label='Constantine (trapezoidal)', color = 'tab:blue' )
plt.plot( Fct_peaks, zct_peaks, 'o', color = 'tab:blue' )

# plt.plot( F, z, label = 'Complet Schijf', color = 'tab:blue' )
# plt.plot( F_peaks, z_peaks, 'o', color = 'tab:blue' )

plt.plot( Fv[z<zv_peaks], z[z<zv_peaks], label = 'Varyani', color = 'orange' )
plt.plot( Fv_peaks, zv_peaks, 'o', color = 'orange' )

plt.plot( Fc[z<zc_peaks], z[z<zc_peaks], label='Constantine (rectangular)', color = 'green' )
plt.plot( Fc_peaks, zc_peaks, 'o', color = 'green' )


plt.plot( Fb[z<zb_peaks], z[z<zb_peaks], label = 'Bouwmeester', color = 'red' )
plt.plot( Fb_peaks, zb_peaks, 'o', color = 'red' )



plt.xlabel( r'Fr' )
plt.ylabel( r'$\dfrac{z}{H_m}$' )


plt.legend()
plt.title(f'{canal}')
plt.savefig( '../figures/Schijf_Complet-z.pdf', bbox_inches = 'tight')
#plt.ylim(0, 0.2)
#plt.xlim(0, 0.2)
plt.show()
```
